```python
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load and explore the Dataset
# Load the dataset
df = sns.load_dataset("tips")
# Display first five rows
print(df.head())
# Check dataset information
print(df.info())
# Summary statistics
print(df.describe())
```

```python
# Distribution of total bill
plt.figure(figsize=(8, 5))
sns.histplot(df["total_bill"], bins=20, kde=True, color="blue")
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot for Tip Amount
plt.figure(figsize=(6, 4))
sns.boxplot(y=df["tip"], color="orange")
plt.title("Box Plot of Tips")
plt.ylabel("Tip Amount ($)")
plt.show()
```

```python
# Scatter plot of total bill vs tip
plt.figure(figsize=(8, 5))
sns.scatterplot(x="total_bill", y="tip", data=df, hue="sex", style="smoker", palette="coolwarm")
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip Amount ($)")
plt.show()
```

```python
# Box plot of Tips by Day
plt.figure(figsize=(8, 5))
sns.boxplot(x="day", y="tip", data=df, palette="Set2")
plt.title("Tip Amounts by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Tip Amount ($)")
plt.show()
```

```python
# Violin plot of Total Bill by Gender
plt.figure(figsize=(8, 5))
sns.violinplot(x="sex", y="total_bill", data=df, palette="husl", inner="quartile")
plt.title("Distribution of Total Bill by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
# Pair Plot for Relationships
sns.pairplot(df, hue="sex", palette="husl")
plt.show()
```

```python
# Heatmap for Correlation Matrix
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
